```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)
df = iris.frame
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
df['sepal_length_group'] = pd.cut(df['sepal_length'], 
                                 bins=3, 
                                 labels=['Short', 'Medium', 'Long'])
```

```python
grouped_stats = df.groupby('sepal_length_group', observed=False)['petal_length'].agg(
    ['mean', 'median', 'min', 'max', 'std']
)
```

```python
print("Summary statistics of Petal Length grouped by Sepal Length Group:\n")
print(grouped_stats)
```

```python
print("\nStatistical details for each species: \n")
species_list = df['species'].unique()

for species in species_list:
    species_data = df[df['species'] == species]
    print(f"\nSpecies: {species}")
    
    print("\nMean:")
    print(species_data.mean(numeric_only=True))
    
    print("\nStandard Deviation:")
    print(species_data.std(numeric_only=True))
    
    print("\nPercentiles (25%, 50%, 75%):")
    print(species_data.quantile([0.25, 0.5, 0.75], numeric_only=True))
    print("-" * 50)
```
